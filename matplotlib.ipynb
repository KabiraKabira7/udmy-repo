## *Implementing figures and axes*

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,10,1)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('polynomial of degree two')
plt.xlim(0,12)
plt.ylim(0,100)
```

```python
fig=plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim()
axes1.set_ylim()
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('power of 4')

#SMALL AXES
axes2=fig.add_axes([0.2,0.2,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
plt.show
```
